## Session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Basic_Operations").getOrCreate()
```

## Read dataset

```python
df = spark.read.csv('appl_stock.csv', inferSchema=True, header=True)

df.printSchema()
```

```python
df.show(5)
```

## Filter

```python
df \
    .filter("Close < 500").show(5)
```

```python
df \
    .filter("Close < 500") \
    .select('Open').show(5)
```

```python
df \
    .filter("Close<500") \
    .select(['Open','Close']).show(5)
```

```python
df \
    .filter(df["Close"] < 200).show(5)
```

```python
df \
    .filter(
        (df["Close"] < 200) &
        ~(df['Open'] > 200)
    ).show(5)
```

## Collect result as python object

```python
df \
    .filter(df["Low"] == 197.16).show()
```

```python
result = df \
    .filter(df["Low"] == 197.16).collect()

result
```
