```python
import pandas as pd
import numpy as np

%pylab inline
```

```python
sample10000 = pd.read_csv('sample10000.csv', parse_dates=[1,2])
sample10000.head()
```

## 1. How many passengers in the sample paid for their ride with cash.

```python
payment_type = (sample10000['payment_type']).value_counts()
payment_type
```

```python
is_cashed = (sample10000['payment_type'] == 2).value_counts()
is_cashed
```

## 2. Build a 99% confidence interval for the proportion of cash payers. What is its' lower boundary?

```python
is_cashed_mean = (sample10000['payment_type'] == 2).mean()
is_cashed_mean
```

```python
is_cashed_std = (sample10000['payment_type'] == 2).std(ddof=1) / np.sqrt(len(sample10000['payment_type']))
is_cashed_std
```

```python
# 1
from statsmodels.stats.weightstats import _tconfint_generic

_tconfint_generic(
    is_cashed_mean, 
    is_cashed_std, 
    len(sample10000['payment_type']) - 1, 
    0.01, 'two-sided')
```

```python
# 2
is_cash = sample10000.payment_type == 2
from statsmodels.stats.proportion import proportion_confint
proportion_confint(sum(is_cash), len(is_cash), alpha=0.01)
```

## 3. Use the same sample to estimate the average trip distance in miles.

```python
trip_distance_mean = sample10000['trip_distance'].mean()
trip_distance_mean
```

## 4. What is the standard deviation of the estimator from the previous question?

```python
trip_distance_std = sample10000['trip_distance'].std(ddof=1) / np.sqrt(len(sample10000['trip_distance']))
trip_distance_std
```

## 5. Calculate 95% confidence interval for the mean trip distance. What is the upper boundary?

```python
from statsmodels.stats.weightstats import _tconfint_generic

_tconfint_generic(
    trip_distance_mean, 
    trip_distance_std, 
    len(sample10000['payment_type']) - 1, 
    0.05, 'two-sided')
```
